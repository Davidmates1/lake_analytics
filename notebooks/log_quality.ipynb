### Select logs

```python
# Filtrando el DataFrame logs con condiciones
selected_logs = logs[
    (logs["exists_true_image"] == True) & 
    (logs["water_with_no_clouds_area"] > 0.1 * logs["total_pixel_area"]) & 
    (logs["cloud_area"] < 0.5 * logs["total_pixel_area"])
]
```

```python
selected_logs
```

```python
selected_logs.osmid.nunique()
```

```python
# 1. Obtener los osmids que están en logs pero no en selected_logs
osmids_diff = set(logs['osmid']) - set(selected_logs['osmid'])

# 2. Filtrar el DataFrame logs para obtener las filas correspondientes a esos osmids y donde exists_true_image es True
filtered_logs = logs[(logs['osmid'].isin(osmids_diff)) & (logs['exists_true_image'] == True)]

# 3. Contar cuántas filas hay para cada valor de osmid en logs
osmid_counts = filtered_logs['osmid'].value_counts()
```

```python
columns_to_use = ["chl_very_low_area", "chl_low_area", "chl_moderate_area", "chl_high_area", "chl_very_high_area"]

weights = np.array([0.2, 0.4, 0.6, 0.8, 1])
selected_logs.loc[:, "mci"] = (selected_logs[columns_to_use].values @ weights) / 1000000
selected_logs.loc[:, "relative_mci"] = selected_logs["mci"] * 1000000 / selected_logs["water_with_no_clouds_area"]

ponderated_weights = np.array([0.1, 0.2, 0.4, 1, 2.5])
selected_logs.loc[:, "ponderated_mci"] = selected_logs[columns_to_use].values @ ponderated_weights / 1000000
selected_logs.loc[:, "relative_ponderated_mci"] = selected_logs["ponderated_mci"] * 1000000 / selected_logs["water_with_no_clouds_area"]
```

```python
selected_logs
```

```python
# 1. Ordenar por 'mci' (de mayor a menor), agrupar por 'osmid' y obtener el mejor registro de cada lago
top_mci_lakes = selected_logs.sort_values('mci', ascending=False).groupby('osmid').first().reset_index().sort_values('mci', ascending=False)

# 2. Quedarse con los 10 primeros lagos con mejor 'mci'
top_10_mci_lakes = top_mci_lakes.nlargest(10, 'mci')[['osmid', 'mci']]

# Repetir el mismo proceso para 'relative_mci'
top_relative_mci_lakes = selected_logs.sort_values('relative_mci', ascending=False).groupby('osmid').first().reset_index().sort_values('relative_mci', ascending=False)
top_10_relative_mci_lakes = top_relative_mci_lakes.nlargest(10, 'relative_mci')[['osmid', 'relative_mci']]

# Y para 'ponderated_mci'
top_ponderated_mci_lakes = selected_logs.sort_values('ponderated_mci', ascending=False).groupby('osmid').first().reset_index().sort_values('ponderated_mci', ascending=False)
top_10_ponderated_mci_lakes = top_ponderated_mci_lakes.nlargest(10, 'ponderated_mci')[['osmid', 'ponderated_mci']]

# También para 'relative_ponderated_mci'
top_relative_ponderated_mci_lakes = selected_logs.sort_values('relative_ponderated_mci', ascending=False).groupby('osmid').first().reset_index().sort_values('relative_ponderated_mci', ascending=False)
top_10_relative_ponderated_mci_lakes = top_relative_ponderated_mci_lakes.nlargest(10, 'relative_ponderated_mci')[['osmid', 'relative_ponderated_mci']]

# Mostrar los resultados
print("Top 10 lagos según MCI:")
print(top_10_mci_lakes)

print("\nTop 10 lagos según Relative MCI:")
print(top_10_relative_mci_lakes)

print("\nTop 10 lagos según Ponderated MCI:")
print(top_10_ponderated_mci_lakes)

print("\nTop 10 lagos según Relative Ponderated MCI:")
print(top_10_relative_ponderated_mci_lakes)

# Convertir a sets para facilitar la intersección
top_mci_ids = set(top_10_mci_lakes['osmid'])
top_relative_mci_ids = set(top_10_relative_mci_lakes['osmid'])
top_ponderated_mci_ids = set(top_10_ponderated_mci_lakes['osmid'])
top_relative_ponderated_mci_ids = set(top_10_relative_ponderated_mci_lakes['osmid'])

# Encontrar la intersección entre los cuatro conjuntos
repeated_lakes = top_mci_ids.intersection(top_relative_mci_ids, top_ponderated_mci_ids, top_relative_ponderated_mci_ids)

# Mostrar los osmids de los lagos que se repiten en los cuatro criterios
print("Lagos que se repiten en los cuatro criterios (MCI, Relative MCI, Ponderated MCI, Relative Ponderated MCI):")
print(repeated_lakes)

# Si también quieres ver los lagos que se repiten en al menos dos criterios:
repeated_in_two_criteria = (top_mci_ids & top_relative_mci_ids) | (top_mci_ids & top_ponderated_mci_ids) | \
                           (top_mci_ids & top_relative_ponderated_mci_ids) | (top_relative_mci_ids & top_ponderated_mci_ids) | \
                           (top_relative_mci_ids & top_relative_ponderated_mci_ids) | (top_ponderated_mci_ids & top_relative_ponderated_mci_ids)

print("\nLagos que se repiten en al menos dos criterios:")
print(repeated_in_two_criteria)
```

```python
all_top_osmids = top_mci_ids.union(top_relative_mci_ids).union(top_ponderated_mci_ids).union(top_relative_ponderated_mci_ids)

filtered_lakes = lakes[lakes['osmid'].isin(all_top_osmids)]

filtered_lakes
```
