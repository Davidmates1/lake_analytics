```python
import math
import matplotlib.pyplot as plt
import numpy as np

def calcular_mci(chl):
    if chl <= 0:
        raise ValueError("El valor de 'chl' debe ser mayor que 0.")
    return 0.0077 * math.log(chl) - 0.017

def calcular_chl(mci):
    if mci<0:
        return 0
    return math.exp((mci + 0.017) / 0.0077)
    

chl_values = np.linspace(0.01, 300, 5000)
resultados = [calcular_mci(mci) for mci in chl_values]

# Crear la grÃ¡fica
plt.plot(chl_values, resultados)
plt.grid(True)
plt.show()
```
